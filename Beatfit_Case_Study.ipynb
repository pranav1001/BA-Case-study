```python
import pandas as pd
import numpy as np
import os
```

```python
pwd =os.getcwd()
```

```python
hourlyCalories=pd.read_csv(pwd+'/drive/MyDrive/hourlyCalories_merged.csv')
hourlyIntensities=pd.read_csv(pwd+'/drive/MyDrive/hourlyIntensities_merged.csv')
hourlySteps=pd.read_csv(pwd+'/drive/MyDrive/hourlySteps_merged.csv')
heartrate_data=pd.read_csv(pwd+'/drive/MyDrive/heartrate_seconds_merged.csv')
```

```python
# Lets clean and manipulate hourlyCalories
new_Calories=hourlyCalories.copy()
new_Calories.dropna(inplace=True)
new_Calories.head()
```

```python
new_Calories['date']=new_Calories['ActivityHour'].str.split(" ",n=1,expand=True)[0]
new_Calories['date']=pd.to_datetime(new_Calories['date'], format='%m/%d/%Y')
new_Calories['weekday']=new_Calories['date'].dt.day_name()
new_Calories['time']=new_Calories['ActivityHour'].str.split(" ",n=1,expand=True)[1]
new_Calories['time']=pd.to_datetime(new_Calories['time'], format='%I:%M:%S %p').dt.strftime('%H:%M')
new_Calories.head()
```

```python
new_Calories['time'].unique()
```

```python
# Lets clean and manipulate hourlyIntensities
new_Intensities=hourlyIntensities.copy()
new_Intensities.dropna(inplace=True)
new_Intensities.head()
```

```python
new_Intensities['date']=new_Intensities['ActivityHour'].str.split(" ",n=1,expand=True)[0]
new_Intensities['date']=pd.to_datetime(new_Intensities['date'], format='%m/%d/%Y')
new_Intensities['weekday']=new_Intensities['date'].dt.day_name()
new_Intensities['time']=new_Intensities['ActivityHour'].str.split(" ",n=1,expand=True)[1]
new_Intensities['time']=pd.to_datetime(new_Intensities['time'], format='%I:%M:%S %p').dt.strftime('%H:%M')
new_Intensities.head()
```

```python
new_Intensities['time'].unique()
```

```python
new_Steps=hourlySteps.copy()
new_Steps.dropna(inplace=True)
new_Steps.head()
new_Steps['date']=new_Steps['ActivityHour'].str.split(" ",n=1,expand=True)[0]
new_Steps['date']=pd.to_datetime(new_Steps['date'], format='%m/%d/%Y')
new_Steps['weekday']=new_Steps['date'].dt.day_name()
new_Steps['time']=new_Steps['ActivityHour'].str.split(" ",n=1,expand=True)[1]
new_Steps['time']=pd.to_datetime(new_Steps['time'], format='%I:%M:%S %p').dt.strftime('%H:%M')
new_Steps.head()
```

```python
heartrate_data.shape
```

```python
heartrate_data.head()
```

```python
new_heartrate=heartrate_data.copy()
new_heartrate.dropna(inplace=True)
```

```python
new_heartrate['date']=new_heartrate['Time'].str.split(" ",n=1,expand=True)[0]
new_heartrate['date']=pd.to_datetime(new_heartrate['date'], format='%m/%d/%Y')
new_heartrate['weekday']=new_heartrate['date'].dt.day_name()
new_heartrate['time_only']=new_heartrate['Time'].str.split(" ",n=1,expand=True)[1]
new_heartrate['time_only']=pd.to_datetime(new_heartrate['time_only'], format='%I:%M:%S %p').dt.strftime('%H:00')
new_heartrate.tail()
```

```python
new_heartrate.shape
```

```python
heartrate_group=new_heartrate.groupby([new_heartrate['Id'],new_heartrate['date'],new_heartrate['time_only']])
```

```python
final_heartrate=heartrate_group['Value'].agg([np.average])
final_heartrate
```

```python
merged_hourly= pd.merge(left=new_Steps,right=final_heartrate,how='left',left_on=['Id','date','time'],right_on=['Id','date','time_only'],right_index=True)
```

```python
merged_hourly
```

```python
merged_hourly=pd.merge(left=new_Calories,right=new_Intensities,how='left',left_on=['Id','date','time'],right_on=['Id','date','time'],suffixes=('','_drop'))
```

```python
merged_hourly=merged_hourly.filter(regex='^(?!.*_drop)')
```
